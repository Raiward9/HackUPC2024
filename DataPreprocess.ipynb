```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
df = pd.read_csv("hackupc-travelperk-dataset-extended.csv")
df
```

```python
df.dtypes
```

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'], dayfirst=True)
df['Return Date'] = pd.to_datetime(df['Return Date'], dayfirst=True)
df
```

Showing cities

```python
cities = list(df.loc[:,'Departure City'].unique())
cities.sort()
print(cities)
```

```python
jsondf = df.to_json(orient='records', date_format='iso')
y = json.loads(jsondf)
```

```python
activities = set()
def insertActivitiesSet(activitiesInstance):
    for activity in activitiesInstance:
        activities.add(activity)
```

```python
for obj in y:
    act = obj['Activities']
    obj['Activities'] = [x.strip() for x in act.split(sep=',')]
    insertActivitiesSet(obj['Activities'])
obj
    
```

```python
activities = list(activities)
activities.sort()
print(activities)
```
