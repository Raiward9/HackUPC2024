```python
import matplotlib.pyplot as plt
import pandas as pd
import json
import random
```

```python
df = pd.read_csv("hackupc-travelperk-dataset.csv")
df
```

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'], dayfirst=True)
df['Return Date'] = pd.to_datetime(df['Return Date'], dayfirst=True)
df
```

```python
jsondf = df.to_json(orient='records')
```

Inputting tags in JSON

```python
def randomTags():
    TAGS = {"hiking", "food_tasting", "sightseeing", "rock_climbing", "museum",
        "music_festival", "wine_tasting", "boating", "camping", "scuba_diving", "bungee_jumping"}
    
    number = random.randint(0,len(TAGS))
    return random.sample(TAGS,number)

data = json.loads(jsondf)

for ind, item in enumerate(data):
    data[ind]['tags'] = randomTags()
    
```

```python
outputJson = json.dumps(data)
with open('travel_data.json', 'w') as outfile:
    outfile.write(outputJson)
```
